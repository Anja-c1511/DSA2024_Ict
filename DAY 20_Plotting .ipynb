### Matplotlib

#### Line Chart

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,400) 
y = np.cos(x)
plt.plot(x,y)
plt.ylim(-0.5,-0.5) # limits the y axia values to the specified limit and zoom in the graph
plt.show()
```

```python
import numpy as np
x = np.arange(0,400) 
y = np.cos(x)
plt.plot(x,y)
plt.xlim(0,20) # Works the same way as ylim 
plt.show()
```

```python
x = [0, 1, 2, 3,4 ,5, 6] 
y = [0,3,4,6,7,8,9]
y1 =[3,4,5,6,7,9,4,]
plt.plot(x,y, c="g",label = "location1") # C specifies the colour of the plot 
plt.plot(x,y1, c="r",label = "location1")
plt.xlabel("days") # plots the x and y axis labels in the graph
plt.ylabel("counts")
plt.legend(loc = "upper left") # we can change the legend location by specifying the position. 
                               # If we do not mention the legent position it will automatically fit in a suitable position. 
plt.show()
```

#### Histogram

```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\kiran\\Downloads\\auto-mpg.csv")
```

```python
# Used to plot graphs of categorical variable ccountries
plt.hist(df["mpg"], color ="blue",rwidth = 0.8, alpha = 0.6, density = True, cumulative = True)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Density")
plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks()
plt.savefig("C:\\Users\\kiran\\OneDrive\\Documents\\Case study\\histogram_chart.png", dpi = 300)
plt.show()
```

#### Scatter plot

```python
# To visualize the correlation between two numerical variables.
plt.figure(figsize=(6,4))
plt.scatter(df["weight"],df["mpg"],c="k",marker ="*",edgecolors="r")
plt.title("Weight Vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

```python
# plotting subplota.
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(df["weight"],df["mpg"],c="k",marker ="*")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.subplot(2,1,2)
plt.scatter(df["displacement"],df["mpg"],c="blue",marker =".")
plt.xlabel("displacement")
plt.ylabel("MPG")
plt.show()
```

#### Box plot

```python
import numpy as np
a = np.random.normal(170,1,20000)
q1 = np.quantile(a,0.25)
q3 = np.quantile(a,0.75)
IQR = q3-q1
lowerbound = q1 - (1.5*IQR)
upperbound = q3 + (1.5*IQR)
sns.boxplot(a)
```
