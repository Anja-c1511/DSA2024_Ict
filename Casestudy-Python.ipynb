### 5. Broadcasting
- Increase all horsepower values by 10% and store the updated values in a new NumPy array. Handle missing data (if any) by replacing it with the mean of the column before applying the increase.


```python
all_Horsepower = np.array(df["horsepower"]) # Getting the horsepower values as a numpy array
Horsepower = np.array(df[df["horsepower"]!= "?"]["horsepower"]) # horsepower values except those values with "?"string
Hrs_pwr = Horsepower.astype(float) # changing the data type to int fronm string
Mean = np.round(np.mean(Hrs_pwr),2) # Mean of all the values except the string "?"
all_Horsepower[all_Horsepower == "?"] = Mean
all_Horsepower.astype(float) # All horsepower values with "?" string replaced by mean value
```

### 10.Conditional Aggregates
- Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
mpg_new = np.array(df["mpg"])
cylinders_new = np.array(df["cylinders"])
unique_cylinders = np.unique(cylinders_new) # Finding unique cylinder groups
unique_cylinders
```

```python
mean_by_cyl = {} # Intializing dictionary
for cyl in unique_cylinders:
    mean_by_cyl[cyl] = np.round(np.mean(mpg_new[cylinders_new == cyl]),3) # Loop for finding mean of mpg for each group of cylinders
print(" Mean mpg by cylinder group:", mean_by_cyl)
    
```

### 10. Finding Anomalies
- Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:
  - Calculate the IQR for mpg.
  - Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 +
1.5 * IQR.
  - Create a DataFrame of cars classified as outliers, displaying car_name, mpg,
and model_year

```python
# Calculate the IQR for mpg.
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR.
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
outliers
```

```python
# Create a DataFrame of cars classified as outliers, displaying car_name, mpg, and model_year
outliers_df = outliers[['car name', 'mpg', 'model year']]
outliers_df
```
