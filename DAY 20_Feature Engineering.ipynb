## Feature Engineering 
- ### Feature Orthogonality
    - It is better if features are orthogonal to each other
    - Orthogonality simply means the features are in opposite direction
    - Selecting those features having extreme correlation with a label to train a model would be better
    - Also using features with extreme correlation among themselves is better.
    - Cosine similarity : dotproduct(u,v)/(|u| |v|), when cosine similarity is zero features are orthogonal. 
- ### Feature Colinearity
    - If a feature is a combination of two other features eg : c = 2a + b 
    - a and b are colinear, so it is undesirable to use colinear features for training.
- ### Feature Slicing
    - Consider a example data set having demographic details(weight,height,haircolur,eyecolour,country) of two countries like india and usa 
    - we need to predict the gender 
    - India has 80% data and usa has 20% data which is not a uniform distribution.
    - Since it is not a uniform distribution it is better to train the data in two different models for the specific countries to predict the gender.-  - ### Feature Binning
    -  For example from the height data of class of students , we can bin the heights into multiple categories
    -  If this is done with only one condition it will be called indicator variable eg: categoring into students having height above and below 150
- ### Mathematical Transforms
    - Logarithams : Can be used to make the tailed distribution to uniform distribution which is better for modelling.


```python
# Box cox and logarithmic transformations

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
data = np.random.exponential(scale =1, size = 10000)
data_log = np.log(data)
transformed_data, lamda_value = stats.boxcox(data)

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.hist(data, bins = 30)
plt.title("Orginal Data")

plt.subplot(1,3,2)
plt.hist(data_log, bins = 30)
plt.title("log Data")

plt.subplot(1,3,3)
plt.hist(transformed_data, bins =30)
plt.title("Box cox transformed data")
plt.show()
```
